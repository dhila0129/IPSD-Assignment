Eksplorasi Data dan Visualisasi

2. Teknik kuantitatif -> metode analisis yang menggunakan angka dan statistik untuk memahami karakteristik dan hubungan dalam data

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Visualisasi
import seaborn as sns # Visualisasi

from sklearn.impute import KNNImputer # Imputasi
from sklearn.preprocessing import StandardScaler # Scaling
from imblearn.over_sampling import SMOTE # Oversampling
from wordcloud import WordCloud # Visualisasi Kata
```

```python
def count_outliers_iqr(data):
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - (1.5 * IQR)
  upper_bound = Q3 + (1.5 * IQR)
  return ((data < lower_bound) | (data > upper_bound)).sum() # Ketika data kurang dari batas bawah atau data lebih dari batas atas, maka dilakukan penambahan

# Count outliers in each numerical column
outlier_counts = {}
for col in df.select_dtypes(include=['int64', 'float64']).columns:
  outlier_counts[col] = count_outliers_iqr(df[col])

# Convert the results into a DataFrame form easier viewing
outlier_counts_df = pd.DataFrame(list(outlier_counts.items()), columns=['Column', 'Outlier Count'])

# Display the outlier counts DataFrame
outlier_counts_df
```

```python
def plot_boxplot(data):
    plt.figure(figsize=(15, 10))
    for i, column in enumerate(data.columns[:-1]):
        plt.subplot(3, 3, i+1)
        sns.boxplot(x='Outcome', y=column, data=data)
        plt.title(f'Box PLot of {column} by Diabetes Outcome')
    plt.tight_layout()
    plt.show
    
plot_boxplot(df)
```

```python
df_text = pd.read_excel('foodreviews.xlsx')
df_text.head(3)
```

```python
text = " ".join(review for review in df_text.Text)

def plot_wordcloud(text):
    wordcloud = WordCloud(width=800, height=400,
                          background_color='white',
                          colormap='viridis').generate(text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud of Review', fontsize=16)
    plt.show
    
plot_wordcloud(text)    
```

# Preproccessing

```python
# Melakukan imputasi dengan KNN
imputer = KNNImputer(n_neighbors=5)
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
x=df.drop('Outcome', axis=1)
y=df['Outcome'] # Menyimpan Outcome ke dalam variabel y
```

```python
smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
sns.countplot(data=x_resampled, y=y_resampled)
```
